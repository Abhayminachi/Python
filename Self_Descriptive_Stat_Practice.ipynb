```python
c=0
for k in ['red','green','yellow','black','blue']:
    c=c+1
    plt.plot(x,x**2+c*2,color=k,label=k)
    plt.legend()
```

```python
col = ['red','green','yellow','black','blue']
```

```python
lab = ['first','sec','third','fourht','fifth']
```

```python
fol = [x**2+1,x**3-10,x**2+3,x**3-5,x**2+5]
```

```python
for m,n,p in zip(fol,lab,col):
    plt.plot(x,m,label=n,color=p)
    plt.legend()
```
