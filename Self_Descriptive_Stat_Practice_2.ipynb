```python
plt.plot(df.PassengerId,df.Age)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(df.PassengerId,df.Age)
ax.set_xlim(100, 500)
ax.set_ylim(0, 80)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(df.PassengerId,df.Age,label = 'Age',color = 'red',linewidth=0.8)
ax.plot(df.PassengerId,df.Fare,label = 'fare',color='black',linestyle = '--')
ax.set_xlim(0,200)
ax.set_title('Plot of Age and fare')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(3,3))
plt.scatter(df.Age,df.Fare,color='blue',s=5)
plt.scatter(df.Age,df.Pclass,color='green',s=4)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age V/s Fare")
plt.ylim(0,30)
plt.ylim(0,80)
plt.legend(['Fare','Class'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(2,2))
plt.bar(df.Pclass,df.Age,width=0.5,color='maroon')
df.groupby('Pclass')['Age'].max()
plt.xlabel("Classes")
plt.ylabel("Max Age in class")
# plt.legend()
plt.show()
```
