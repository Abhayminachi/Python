```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
fig,ax = plt.subplots()

bar_labels = ['Kempu', 'Neeli', 'Gulabi', 'Kesari']
bar_colors = ['red', 'blue', 'pink', 'green']
ax.bar(fruits,counts,label = bar_labels,color = bar_colors)
ax.legend(title = bar_labels)
plt.show()
```

```python
plt.bar(fruits,counts,label = bar_labels, color = ['red','blue','yellow','green'],edgecolor = 'blue')
plt.legend()
plt.show()
```
